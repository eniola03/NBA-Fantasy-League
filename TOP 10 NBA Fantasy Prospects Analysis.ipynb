```python
#import necessary libraries
import pandas as pd
import numpy as np
```

```python
#bringing in the following data
Regular_ssn = pd.read_csv('2021-2022 NBA Player Stats - Regular.csv', delimiter=";", encoding='latin-1')

Regular_ssn.head(10)
```

Now that we've visualized the following dataframe we can begin to analyze the data, though first and foremost I'll have to add an extra column to our dataframe that calculates the players efficiency with the given formula, EFF = (PTS + REB + AST + STL + BLK − Missed FG − Missed FT - TO) / GP.

```python
#First need to see the column names to accesss easily
Regular_ssn.columns 
```

```python
Regular_ssn['EFF'] = (((Regular_ssn.PTS + Regular_ssn.ORB + Regular_ssn.DRB + Regular_ssn.AST + Regular_ssn.STL + 
                      Regular_ssn.BLK) - (Regular_ssn.FGA - Regular_ssn.FG) - (Regular_ssn.FTA - Regular_ssn.FT) - Regular_ssn.TOV) / Regular_ssn.G) * 100
```

```python
Regular_ssn.head()
#we can see that the new column was implemented
```

Looking for Promising "Stars" within the league: By looking at a dataframe of players who score at least 15 points a game with a minimum of 50 games played

```python
Cs = pd.DataFrame(Centers.Player[(Centers.Age<=38)&(Centers.PTS>=15)&(Centers.EFF>=30.0)&(Centers.G > 50)]).rename(columns={"Player": "Centers"}).reset_index(drop=True)

PGs = pd.DataFrame(Point_Guards.Player[(Point_Guards.Age<=38)&(Point_Guards.PTS>=15)&(Point_Guards.EFF>=30.0)&(Point_Guards.G > 50)]).rename(columns={"Player": "Point Guards"}).reset_index(drop=True)

SGs = pd.DataFrame(Shooting_Guards.Player[(Shooting_Guards.Age<=38)&(Shooting_Guards.PTS>=15)&(Shooting_Guards.EFF>=30.0)&(Shooting_Guards.G > 50)]).rename(columns={"Player": "Shooting Guards"}).reset_index(drop=True)

PFs = pd.DataFrame(Power_Forward.Player[(Power_Forward.Age<=38)&(Power_Forward.PTS>=15)&(Power_Forward.EFF>=30.0)&(Power_Forward.G > 50)]).rename(columns={"Player": "Power Forwards"}).reset_index(drop=True)

SFs = pd.DataFrame(Small_Forward.Player[(Small_Forward.Age<=38)&(Small_Forward.PTS>=15)&(Small_Forward.EFF>=30.0)&(Small_Forward.G > 50)]).rename(columns={"Player": "Small Forwards"}).reset_index(drop=True)

Stars = pd.concat([Cs, PGs, SGs, PFs, SFs], axis=1, join="outer")
Stars.head(15)
```
